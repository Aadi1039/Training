```python
import numpy as np
import pandas as pd

data={
    'Aditya':pd.Series([80,90,91],index=['M1','M2','M3']),
    'Kriti':pd.Series([80,90,91,92],index=['M1','M2','M3','M4']),
    'Krishu':pd.Series([80,81,82],index=['M1','M2','M3']),
    'kouser':pd.Series([80,90,91,92],index=['M1','M2','M3','M4'])
}
df=pd.DataFrame(data)
print(df)
#deleting a column
del df['Kriti']
print(df)
df.pop('kouser')
print(df)
del df['Krishu']
print(df)
```
