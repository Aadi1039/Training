```python
import apache_beam as beam

def process_string(word):
    return f"{word.upper()} (length={len(word)})"

with beam.Pipeline() as p:
    (
        p
        | "String" >> beam.Create([
            "Aditya",
            "Kumar",
            "CSE",
            "6 SEM",
            "python"
        ])
        | "Strings" >> beam.Map(process_string)
        | "Output" >> beam.Map(print)
    )
```
