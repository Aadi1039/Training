```python
a = 20
b = 5

print(f"a = {a}, b = {b}")
print(f"a + b = {a + b}")
print(f"a - b = {a - b}")
print(f"a * b = {a * b}")
print(f"a / b = {a / b}") # Float division
print(f"a % b = {a % b}") # Modulus
print(f"a ** b = {a ** b}") # Exponentiation
print(f"a // b = {a // b}") # Floor division
```

```python
a = 20
b = 5

print(f"a = {a}, b = {b}")

print(f"a == b : {a == b}")
print(f"a != b : {a != b}")
print(f"a > b  : {a > b}")
print(f"a < b  : {a < b}")
print(f"a >= b : {a >= b}")
print(f"a <= b : {a <= b}")
```

```python
a = 10
b = 20

print("--- Logical Operators ---")
x_bool = bool(a) # True since a is non-zero
y_bool = bool(b) # True since b is non-zero
z_bool = False   # A third boolean for demonstration

print(f"a = {a}, bool(a) = {x_bool}")
print(f"b = {b}, bool(b) = {y_bool}")
print(f"z_bool = {z_bool}")

print(f"bool(a) and bool(b) : {x_bool and y_bool}") # True and True = True
print(f"bool(a) or z_bool : {x_bool or z_bool}")   # True or False = True
print(f"not bool(a) : {not x_bool}")           # not True = False
print(f"not z_bool : {not z_bool}")           # not False = True
```

```python
# Conditional Operator (Ternary Operator)

a = 10
b = 5

# Syntax: value_if_true if condition else value_if_false
result = "a is greater" if a > b else "b is greater or equal"
print(f"a = {a}, b = {b}")
print(f"Result: {result}")

a = 5
b = 10
result = "a is greater" if a > b else "b is greater or equal"
print(f"\na = {a}, b = {b}")
print(f"Result: {result}")

# Example with equality
c = 7
d = 7
status = "Equal" if c == d else "Not Equal"
print(f"\nc = {c}, d = {d}")
print(f"Status: {status}")
```
